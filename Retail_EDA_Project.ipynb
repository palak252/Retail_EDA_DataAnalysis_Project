```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
top10 = df.sort_values(by="Calories",ascending=False).head(10)
plt.barh(top10["Item"], top10["Calories"])
plt.title("Top 10 highest calorie items")
plt.xlabel("Calories")
plt.show()
```

```python
plt.hist(df["Calories"],bins=20)
plt.title("Calories Distrubution")
plt.xlabel("Calories")
plt.ylabel("Frequency")
plt.show()
```

```python
df.groupby("Category")["Calories"].mean()
```

```python
df ["Category"].value_counts().plot(kind="bar")
plt.title("Numbers of item in each category")
plt.show()
```

```python
plt.scatter(df["Calories"], df["Protein"])
plt.xlabel("Calories")
plt.ylabel("Protein")
plt.title("Calories VS Protein")
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True),annot=True)
plt.title("Correlation Between Nutruitional Values")
plt.show()
```
